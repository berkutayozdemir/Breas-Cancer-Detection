# Breast Cancer Detection with Logistic Regression

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("data.csv") # import the data 
```

```python
df.columns
```

## Data Cleaning & EDA 

```python
df.describe()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull())
```

```python
df.drop(["Unnamed: 32","id"],axis=1,inplace=True)
```

```python
df
```

```python
df.diagnosis = [1 if value == "M" else 0 for value in df.diagnosis]
```

```python
df["diagnosis"] = df["diagnosis"].astype("category",copy=False)
```

```python
df.dtypes
```

```python
df["diagnosis"].value_counts().plot(kind="bar")
```

## Preprocessing

```python
X = df.drop(["diagnosis"],axis=1) # Predictors
y = df["diagnosis"] # Target
```

```python
# Scaling the data 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Spliting Data As Training & Test

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3,random_state=42) 
```

## Training the Model

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
model = lr.fit(X_train,y_train)

y_predicted = lr.predict(X_test)
```

```python
y_predicted
```

```python
y_test
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
accuracy = accuracy_score(y_test,y_predicted)
print(f"Accuracy : {accuracy:.5f}")
```

```python
print(classification_report(y_test,y_predicted))
```

```python
from sklearn.model_selection import learning_curve
```

```python
train_sizes, train_scores, test_scores = learning_curve(
    model, X_train, y_train, cv=5, n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0,8,1000)
)
```

```python
plt.figure()
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', color='r', label='Training score')
plt.plot(train_sizes, np.mean(test_scores, axis=1), 'o-', color='g', label='Cross-validation score')
plt.xlabel('Training Size')
plt.ylabel('Score')
plt.title('Learning Curve for Logistic Regression')
plt.legend()
plt.grid()
plt.show()
```
